```python
import numpy                       
from matplotlib import pyplot                 
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
nx = 41  # try changing this number from 41 to 81 and Run All ... what happens?
dx = 2/(nx-1)
nt = 25    
dt = .02  
c = 1.      #assume wavespeed of c = 1
x = numpy.linspace(0,2,nx)
```

```python
u = numpy.ones(nx)      #numpy function ones()
lbound = numpy.where(x >= 0.5)
ubound = numpy.where(x <= 1)
```

```python
bounds = numpy.intersect1d(lbound, ubound)
u[bounds]=2  #setting u = 2 between 0.5 and 1 as per our I.C.s
```

```python
pyplot.figure(figsize=(4,3))   #set plot size
pyplot.ylim(0,2.5)             #y-axis plot limits
pyplot.tick_params(axis='both', labelsize=14) #increase font size for ticks
pyplot.xlabel('x', fontsize=14) #x label
pyplot.ylabel('u', fontsize=14) #y label
pyplot.plot(x,u, 'k-', ls='--');
```

```python
for n in range(1,nt):  
    un = u.copy() 
    for i in range(1,nx): 
    
        u[i] = un[i]-c*dt/dx*(un[i]-un[i-1])
```

```python
pyplot.figure(figsize=(4,3))   #set plot size
pyplot.ylim(0,2.5)             #y-axis plot limits
pyplot.tick_params(axis='both', labelsize=14) #increase font size for ticks
pyplot.xlabel('x', fontsize=14) #x label
pyplot.ylabel('u', fontsize=14) #y label
pyplot.plot(x,u, 'k-', ls='--');
```

```python
for n in range(1, nt-15):  
    un = u.copy() 
    u[1:] = un[1:]-un[1:]*dt/dx*(un[1:]-un[0:-1]) 
    u[0] = 1.0
```

```python
pyplot.plot(x, u, color='#003366', ls='--', lw=3)
pyplot.ylim(0,2.5);
```
